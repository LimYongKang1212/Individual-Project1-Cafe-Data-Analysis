# **2. Data Preprocessing** #

Data Cleaning

```python
# show all unique value for all column except Transaction ID and Transaction date
for col in df.columns:
  if col not in ['Transaction ID', 'Transaction Date']:
    print(f"{col}: {df[col].unique()}")
```

```python
# replace value "ERROR" AND "UNKNOWN" With null value for better data cleaning
df = df.replace(to_replace=["ERROR", "UNKNOWN"], value= pd.NA)
```

```python
# Change data type

df['Item'] = df['Item'].astype('category')

df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce').astype('Int64')

df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')
df['Total Spent'] = pd.to_numeric(df['Total Spent'], errors='coerce')
df['Payment Method'] = df['Payment Method'].astype('category')
df['Location'] = df['Location'].astype('category')
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], dayfirst=True, format='mixed',errors='coerce')

print(df.info())
df.describe(include="all")
```

```python
# Check missing value
df.isnull().sum()
```

```python
# Fill/replace missing value
# Replace missing value by mode for each category column
df['Item'].fillna(df['Item'].mode()[0], inplace=True)
df['Payment Method'].fillna(df['Payment Method'].mode()[0], inplace=True)
df['Location'].fillna(df['Location'].mode()[0], inplace=True)

# fill in the price per unit based on the menu given 
df.loc[df['Item'] == 'Coffee', "Price Per Unit" ] = 2
df.loc[df['Item'] == 'Tea', "Price Per Unit" ] = 1.5
df.loc[df['Item'] == 'Sandwich', "Price Per Unit" ] = 4
df.loc[df['Item'] == 'Salad', "Price Per Unit" ] = 5
df.loc[df['Item'] == 'Cake', "Price Per Unit" ] = 3
df.loc[df['Item'] == 'Cookie', "Price Per Unit" ] = 1
df.loc[df['Item'] == 'Smoothie', "Price Per Unit" ] = 4
df.loc[df['Item'] == 'Juice', "Price Per Unit" ] = 3

print(df)
print(df.isna().sum())
```

```python
# Replace quantity with total spent divided by price per unit
df['Quantity'] = df['Quantity'].fillna((df['Total Spent'] / df['Price Per Unit']).round().astype('Int64'))

# Replace Total spent with Quantity times Price per unit
df['Total Spent'] = df['Total Spent'].fillna(df['Quantity'] * df['Price Per Unit'])

df.isna().sum()
```

```python
# chcek the column quantity and total spent that consists missing value
df[df['Quantity'].isna() | df['Total Spent'].isna()]
```

```python
# Check mean, median for quantity column and choose the suitable one to replace the missing value
Qtymean = df['Quantity'].mean()
Qtymedian = df['Quantity'].median()

print(Qtymean)
print(Qtymedian)
```

```python
# replace missing value in quantity column with median due to quantity column is int type data
df['Quantity'] = df['Quantity'].fillna(Qtymedian)

# fill na value again for the total spent column  (Quantity*Price Per Unit)
df['Total Spent'] = df['Total Spent'].fillna(df['Quantity'] * df['Price Per Unit'])

df.isna().sum()
```

```python
# check if the data is correct (Quantity * Price per unit  = total Spent)
df[df['Total Spent'] != (df['Quantity'] * df['Price Per Unit'])]

# correct the total spent with replace the true total spent
df.loc[df['Total Spent'] != (df['Quantity'] * df['Price Per Unit']), 'Total Spent'] = df['Quantity'] * df['Price Per Unit']

# recheck again the incorrect total spent
df[df['Total Spent'] != (df['Quantity'] * df['Price Per Unit'])]
```

```python
# check total missing value for transaction date by total data
naDate = (df['Transaction Date'].isna().sum()/df['Transaction Date'].count()*100)
print(naDate)
```

```python
# the total missing value of Transaction date is 5% of the total number of data so  will decided to drop the data directly
df = df.dropna(subset=['Transaction Date'])
print(df.isna().sum())
print(df.shape)
```

```python
# check duplicate data
print(df.duplicated().sum())
```

# **3. Feature Engineering** #

```python
# Create new column for year,month, day and weekday
df["Day"] = df["Transaction Date"].dt.day
df["Month"] = df["Transaction Date"].dt.month_name()
df["Year"] = df["Transaction Date"].dt.year
df["Weekday"] = df["Transaction Date"].dt.day_name()

df
```
